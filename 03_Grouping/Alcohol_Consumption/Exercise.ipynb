### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

```python
url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
```

### Step 3. Assign it to a variable called drinks.

```python
df_drinks=pd.read_csv(url)
df_drinks
```

### Step 4. Which continent drinks more beer on average?

```python
df_beerP=df_drinks.sort_values(by='beer_servings',ascending=False)
df_beerP
```

```python
df_beerP.reset_index().query('country == "Spain"')
```

```python
df_g=df_drinks.groupby('continent')
df_g
```

```python
df_eu=df_g.get_group('EU')
df_eu
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=df_eu.beer_servings,y=df_eu.wine_servings, label='EU')
plt.legend()
```

```python
df_af=df_g.get_group('AF')
df_af
```

```python
plt.scatter(x=df_eu.beer_servings,y=df_eu.wine_servings, label='EU')
plt.scatter(x=df_af.beer_servings,y=df_af.wine_servings, label='AF')
plt.legend()
```

```python
df_drinks.fillna(value="NA",inplace=True)
continentes=df_drinks.continent.unique()
continentes 
```

```python
for continente in continentes:
    df_c=df_g.get_group(continente)
    plt.scatter(x=df_c.beer_servings,y=df_c.wine_servings, label=continente)
plt.legend()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
df_eu=df_g.get_group('EU')
plt.bar(x=df_eu.country,height=df_eu.wine_servings)
```

```python
for continente in continentes:
    df_c=df_g.get_group(continente)
    plt.bar(x=continente,height=df_c.wine_servings)
```

```python
df_drinks.groupby('continent').mean()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
df_drinks.groupby('continent').median()
```
